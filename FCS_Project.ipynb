**6. For each station, determine the buses that have stopped there at least once.**

**9. For each (route, bus) pair, compute the ratio between the overall energy consumption and the overall driven distance.**

```python
short_trips['energy_distance_ratio'] = short_trips['energyConsumption'] / short_trips['drivenDistance']

# Raggruppamento per (busRoute, busNumber) per calcolare il consumo energetico totale e la distanza totale percorsa
grouped_ratio = short_trips.groupby(['busRoute', 'busNumber'])[['energyConsumption', 'drivenDistance']].sum()
grouped_ratio['overall_energy_distance_ratio'] = grouped_ratio['energyConsumption'] / grouped_ratio['drivenDistance']

# Reset dell'indice per avere busRoute e busNumber come colonne
grouped_ratio_reset = grouped_ratio.reset_index()

grouped_ratio_reset.head()
```

<span style="font-size:12px;">Calcolo del Rapporto Energetico/Distanza per Viaggio: tra consumo di energia e distanza percorsa. 
Poi Raggruppiamo per Coppia (Rotta, Autobus).
Calcolo del Rapporto Energetico/Distanza Totale: Per ciascun gruppo (cio per ogni coppia rotta-autobus),si calcola il rapporto tra il consumo energetico totale e la distanza totale percorsa.
Reset dell'Indice: L'indice viene resettato per convertire busRoute e busNumber da indici a colonne normali nel DataFrame grouped_ratio_reset. </span>

**10. Starting from the results of the previous point, for each route compute the buses with max and min energy ratio, and save the difference between these ratios in a dataframe.**

```python
# Trovare gli autobus con il rapporto energetico massimo e minimo per ogni percorso
max_ratio_per_route = grouped_ratio_reset.groupby('busRoute')['overall_energy_distance_ratio'].idxmax()
min_ratio_per_route = grouped_ratio_reset.groupby('busRoute')['overall_energy_distance_ratio'].idxmin()

# Estrazione dei dati per gli autobus con il rapporto massimo e minimo
max_ratio_buses = grouped_ratio_reset.loc[max_ratio_per_route].set_index('busRoute')
min_ratio_buses = grouped_ratio_reset.loc[min_ratio_per_route].set_index('busRoute')

# Calcolo della differenza dei rapporti e salvataggio in un dataframe
difference_df = pd.DataFrame({
    'Route': max_ratio_buses.index,
    'MaxRatioBus': max_ratio_buses['busNumber'],
    'MinRatioBus': min_ratio_buses['busNumber'],
    'MaxRatio': round(max_ratio_buses['overall_energy_distance_ratio'],2),
    'MinRatio': round(min_ratio_buses['overall_energy_distance_ratio'],2),
    'Difference': round(max_ratio_buses['overall_energy_distance_ratio'] - min_ratio_buses['overall_energy_distance_ratio'],2)
}).reset_index(drop=True)
```

**14. For each month of the year, compute the average ambient temperature**

```python
from glob import glob

# Ottieni l'elenco di tutti i file CSV nella cartella
file_paths = glob(os.path.join(directory, '*.csv'))

# Preparazione di un DataFrame per raccogliere i dati aggregati
monthly_temps = pd.DataFrame()

for file_path in file_paths:
    # Estrai il nome del file dal percorso completo
    file_name = os.path.basename(file_path)
    
    # Controlla se il nome del file corrisponde a quello da escludere
    if file_name == 'B208_2022-12-10_23-34-46_2022-12-11_03-29-05.csv':
        continue
    
    df = pd.read_csv(file_path)
    
    # Assumi che 'time_iso' sia la colonna temporale e 'temperature_ambient' quella della temperatura
    # Converti la colonna temporale in formato datetime di pandas
    df['time_iso'] = pd.to_datetime(df['time_iso'])
    
    # Estrai il mese dall'orario e aggiungi come nuova colonna
    df['month'] = df['time_iso'].dt.month
    
    # Calcola la media mensile della temperatura per il file corrente
    monthly_avg = df.groupby('month')['temperature_ambient'].mean().reset_index()
    
    # Combina i risultati con quelli dei file precedenti
    monthly_temps = pd.concat([monthly_temps, monthly_avg], ignore_index=True)

# Calcola la media globale della temperatura per ogni mese su tutti i file
global_monthly_avg = monthly_temps.groupby('month')['temperature_ambient'].mean().reset_index()

global_monthly_avg['temperature_ambient'] = global_monthly_avg['temperature_ambient'].round(2)

# Visualizza i risultati
global_monthly_avg
```

**15. For each bus compute the total time when the halt brake is active and the total time when the park brake is active. Compute also the ratio between those two times.**

```python
 import pandas as pd
from math import radians, cos, sin, asin, sqrt

# Funzione per calcolare la distanza haversiana tra due punti sulla terra
def haversine(lon1, lat1, lon2, lat2):
    # Converti le coordinate decimali in radianti
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    
    # Formula haversiana
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    
    # Raggio della Terra in chilometri (puoi cambiare questo valore per miglia)
    r = 6371
    
    return c * r
```
